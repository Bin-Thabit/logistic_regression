```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df
```

```python
# drop unnecesserly data 
df.drop(['id' , "Unnamed: 32"] , axis = 1, inplace = True)
df
```

```python
# replace each M with 1 and each D with 0
df.diagnosis = [1 if value == 'M' else 0 for value in df.diagnosis]
df['diagnosis']
```

```python
# Get the feature and the target
feature = df.drop(['diagnosis'], axis = 1).to_numpy()
target = df['diagnosis'].to_numpy()
```

```python
# Create the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def loss_function(x, y, w, b):
    m, n = x.shape
    cost = 0.0
    epsilon = 1e-10  # Small value to avoid log(0)
    
    for i in range(m):
        z_i = np.dot(w, x[i]) + b
        model = sigmoid(z_i)
        # Clip the probabilities to avoid log(0)
        model = np.clip(model, epsilon, 1 - epsilon)
        cost += -y[i] * np.log(model) - (1 - y[i]) * np.log(1 - model)
    
    cost /= m
    return cost
```

```python
# Implement the derivative of w and b 
def compute_gradient(x, y, w, b):
    m, n = x.shape
    der_w = np.zeros(n)
    der_b = 0
    for i in range(m):
        z_i = z_i = np.dot(w, x[i]) + b
        model = sigmoid(z_i)
        error = model - y[i]
        for j in range(n):
            der_w[j] += error * x[i][j]
        der_b += error
    der_w /= m
    der_b /= m
    return der_w, der_b
```

```python
# Implement the gradient_descent algorithm:
def gradient_descent(x, y, alpha, num_iter):
    m, n = x.shape
    w = np.zeros(n)
    b = 0.0
    for i in range(num_iter):
        der_w, der_b = compute_gradient(x, y, w, b)
        w -= alpha * der_w
        b -= alpha * der_b

        if i % 100 == 0:
            cost = loss_function(x, y, w, b)
            print(f"Iteration {i:4d}: Cost {cost}")
    return w, b
```

```python
# Normalize both x and y
feature_mean = np.mean(feature, axis=0)
feature_std = np.std(feature, axis=0)


feature_normalize = (feature - feature_mean) / feature_std


# Initialize alpha and number of iteration
alpha = 0.09
num_iter = 20000
w, b = gradient_descent(feature_normalize, target, alpha, num_iter)
```

```python
# Implement the predict function
def predict(x, w, b):
    z = np.dot(w, x) + b
    return sigmoid(z)
```

```python
target_prediction = [1 if predict(feature_normalize[i], w, b) >= 0.5 else 0 for i in range(feature_normalize.shape[0])]
# This variable to count how many mistake 
err = 0
for i in range(feature_normalize.shape[0]):
    print(f"Original: {target[i]} -- Prediction: {target_prediction[i]}")
    if target[i] != target_prediction[i]:
        err += 1
print(err)
print(target.shape)
```
